### (c) v

```python
def plotHists (cols, myBins=[10, 20, 50, 100]):
    rows = len(cols)
    i = 1
    fig=plt.figure(figsize=(12,9))
    for c in cols:
        for j in range(len(myBins)):
            ax=fig.add_subplot(rows,4,i)
            df[c].hist(bins=myBins[0],ax=ax)
            i+=1

    fig.tight_layout()
    
plotHists(['Apps', "Accept","Enroll"])
```
